```python
from datasets import load_dataset
```

```python
raw_datasets = load_dataset("conll2003")
```

```python
print(raw_datasets["train"][4]["tokens"])
print(raw_datasets["train"][4]["ner_tags"])
print([raw_datasets["train"].features["ner_tags"].feature.names[i] for i in raw_datasets["train"][4]["ner_tags"]])
```

```python
from transformers import AutoTokenizer

checkpoint="bert-base-cased"

tokenizer=AutoTokenizer.from_pretrained(checkpoint)
```

```python
inputs=tokenizer(raw_datasets["train"][4]["tokens"], is_split_into_words=True)
print(inputs.tokens())
print(raw_datasets["train"][4]["tokens"])
```

```python
print(inputs.word_ids())
```

```python
def align_labels_with_tokens(labels, word_ids):
    new_labels = []
    current_word = None
    for word_id in word_ids:
        if word_id != current_word:
            # Start of a new word!
            current_word = word_id
            label = -100 if word_id is None else labels[word_id]
            new_labels.append(label)
        elif word_id is None:
            # Special token
            new_labels.append(-100)
        else:
            # Same word as previous token
            label = labels[word_id]
            # If the label is B-XXX we change it to I-XXX
            if label % 2 == 1:
                label += 1
            new_labels.append(label)

    return new_labels
```

```python
labels=raw_datasets["train"][4]["ner_tags"]
print(labels)
print(align_labels_with_tokens(labels, inputs.word_ids()))
```

```python
def tokenize_and_align_labels(examples):
    tokenized_inputs = tokenizer(
        examples["tokens"], truncation=True, is_split_into_words=True
    )
    all_labels = examples["ner_tags"]
    new_labels = []
    for i, labels in enumerate(all_labels):
        word_ids = tokenized_inputs.word_ids(i)
        new_labels.append(align_labels_with_tokens(labels, word_ids))

    tokenized_inputs["labels"] = new_labels
    return tokenized_inputs

tokenized_datasets = raw_datasets.map(
    tokenize_and_align_labels,
    batched=True,
    remove_columns=raw_datasets["train"].column_names,
)
```

```python
from transformers import DataCollatorForTokenClassification

data_collator = DataCollatorForTokenClassification(
    tokenizer=tokenizer, return_tensors="tf"
)
```

```python
tf_train_dataset =tokenized_datasets["train"].to_tf_dataset(
   columns=["attention_mask", "input_ids", "labels", "token_type_ids"],
    collate_fn=data_collator,
    shuffle=True,
    batch_size=16,
) 

tf_validation_dataset =tokenized_datasets["validation"].to_tf_dataset(
   columns=["attention_mask", "input_ids", "labels", "token_type_ids"],
    collate_fn=data_collator,
    shuffle=False,
    batch_size=16,
) 
```

```python
label_names=raw_datasets["train"].features["ner_tags"].feature.names
id2label = {i: label for i, label in enumerate(label_names)}
label2id = {v: k for k, v in id2label.items()}
```

```python
from transformers import TFAutoModelForTokenClassification

model=TFAutoModelForTokenClassification.from_pretrained(checkpoint, id2label=id2label, label2id=label2id)
```

```python
model.config.num_labels
```

```python
from huggingface_hub import notebook_login

notebook_login()
```

```python
from transformers import create_optimizer
import tensorflow as tf

num_epochs = 3
num_train_steps = len(tf_train_dataset) * num_epochs

optimizer, schedule = create_optimizer(
    init_lr=2e-5,
    num_warmup_steps=0,
    num_train_steps=num_train_steps,
    weight_decay_rate=0.01,
)
model.compile(optimizer=optimizer)
```

```python
from transformers.keras_callbacks import PushToHubCallback

callback = PushToHubCallback(output_dir="bert-finetuned-ner", tokenizer=tokenizer)

model.fit(
    tf_train_dataset,
    validation_data=tf_validation_dataset,
    callbacks=[callback],
    epochs=num_epochs,
)
```

```python
import evaluate
import numpy as np


metric = evaluate.load("seqeval")


all_predictions = []
all_labels = []
for batch in tf_validation_dataset:
    logits = model.predict_on_batch(batch)["logits"]
    labels = batch["labels"]
    predictions = np.argmax(logits, axis=-1)
    for prediction, label in zip(predictions, labels):
        for predicted_idx, label_idx in zip(prediction, label):
            if label_idx == -100:
                continue
            all_predictions.append(label_names[predicted_idx])
            all_labels.append(label_names[label_idx])
metric.compute(predictions=[all_predictions], references=[all_labels])
```
